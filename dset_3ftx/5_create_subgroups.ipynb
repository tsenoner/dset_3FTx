```python
import numpy as np
import pandas as pd
from pathlib import Path

base = Path("../data/online_version")
csv_file = base / "20230330_3FTx.csv"
rostspace_csv = base / "rostspace_data.csv"
submission_csv = base / "submission_data.csv"
internal_csv = base / "internal_data.csv"
sfet_csv = base / "sfet_data.csv"
# tmp_out = base / "tmp.csv"
itol_out = base.parent / "phylo_tree/iTOL_group.txt"
```

```python
df = pd.read_csv(csv_file)  # create the groups
df = df.rename(columns={"membran_prediction": "membrane_prediction"})
df["group"] = np.where(df["major_group"] == "Ly-6", df["sub_group"], df["major_group"])
# --- data for SFET ---
sfet_cols = [
    "identifier",
    "group",
    "membrane_prediction",
    "taxon_of_interest",
    "family",
    "genus",
    "species",
    "taxon_id",
    "db",
    "uniprot_id",
    "uniprot_entry",
    "genbank_id",
    "refseq_id",
    "gi_number",
    "activity",
    "seq_start",
    "full_seq",
    "mature_seq",
    "Dimeric",
]
df[sfet_cols].to_csv(sfet_csv, index=False)

# add `in_tree` column (boolen telling if entry is displayd in trees)
df.loc[df["id_new"] < 1417, "in_tree"] = "yes"
df.loc[df["id_new"] >= 1417, "in_tree"] = "no"
# crate activity group
df["activity"] = np.where(
    pd.notna(df["activity"]),
    df["activity"],
    np.where(
        pd.notna(df["zhang_subfamily"]),
        df["zhang_subfamily"],
        np.where(pd.notna(df["ritu_class"]), df["ritu_class"], np.nan),
    ),
)
columns2keep = [
    "identifier",
    "group",
    "major_group",
    "sub_group",
    "membrane_prediction",
    "taxon_numerical",
    "taxon_of_interest",
    "family",
    "genus",
    "species",
    "taxon_id",
    "data_origin",
    "db",
    "uniprot_id",
    "uniprot_entry",
    "genbank_id",
    "refseq_id",
    "gi_number",
    "in_tree",
    "seq_start",
    "full_seq",
    "mature_seq",
    "number_cysteines",
    "activity",
    "Dimeric",
]
df = df[columns2keep]
df.to_csv(internal_csv, index=False)

rostspace_cols = [
    "identifier",
    "group",
    "major_group",
    "taxon_of_interest",
    # "sub_group",
    "membrane_prediction",
    "seq_start",
    # "number_cysteines",
    # "Dimeric"
    # "activity",
]
df[rostspace_cols].to_csv(rostspace_csv, index=False)

submission_cols = [
    "identifier",
    "group",
    "major_group",
    "membrane_prediction",
    "taxon_of_interest",
    "family",
    "genus",
    "species",
    "taxon_id",
    "db",
    "uniprot_id",
    "uniprot_entry",
    "genbank_id",
    "refseq_id",
    "gi_number",
    "in_tree",
    "seq_start",
    "full_seq",
    "mature_seq",
]
df[submission_cols].to_csv(submission_csv, index=False)
```
